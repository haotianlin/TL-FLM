```python
import time 
import datetime
import pandas as pd
```

```python
# single stock
ticker = 'TSLA'
period1 = int(time.mktime(datetime.datetime(2021,4,1,23,59).timetuple()))
period2 = int(time.mktime(datetime.datetime(2021,9,30,23,59).timetuple()))
interval = '1d' # 1wk, 1m

query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'

df = pd.read_csv(query_string)
df['Name'] = 'TSLA'
df.to_csv('TSLA.cvs')


# multiple stocks
tickers = {'TSLA', 'TWTR', 'AAPL', 'GOOG'}
period1 = int(time.mktime(datetime.datetime(2020,12,1,23,59).timetuple()))
period2 = int(time.mktime(datetime.datetime(2020,12,31,23,59).timetuple()))
interval = '1d' # 1wk, 1m

df_all = {}

for ticker in tickers:
    query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'
    df = pd.read_csv(query_string)
    df['Name'] = ticker
    df_all[ticker] = df


dataframes = [ df_all[ticker] for ticker in tickers ]
    
result = pd.concat(dataframes)

result.to_csv('target.cvs')
```
